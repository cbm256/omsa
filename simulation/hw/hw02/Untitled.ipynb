```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
def dist(seed, n):
    d = []
    xi = seed
    for i in range(n+1):
        xi = 16807*(xi)%(2**31-1)
        prn = xi/(2**31-1)
        d.append(prn)
        seed -= 1
        
    return d

    
```

```python
u = dist(3245, 1000)
X = [-np.log(x) for x in u]
sns.distplot(X)
```

```python
seedx = 23542335
seedy = 98624

x = dist(seedx, 1000)
y = dist(seedy, 1000)

z = [np.sqrt(-2*np.log(X))*np.sin(2*np.pi*Y) for X, Y in zip(x,y)]
```

```python
sns.distplot(z)
```

```python
z = [np.sqrt(-2*np.log(X))*np.cos(2*np.pi*Y) for X, Y in zip(x,y)]
sns.distplot(z)
```

```python
Z = []
seedx = 23542335
seedy = 98624647


for i in range(1001):
    x = dist(seedx, 1000)
    y = dist(seedy, 1000)
    z = [X/(X-Y) for X,Y in zip(x,y)]
    Z.append(np.mean(z))
    
    seedx += 1
    seedy -= 1
```

```python
sns.distplot(z)
```

```python
import random

prob = []
for i in range(0,1000):
    di1 = np.array([random.randint(1,6) for x in range(1001)])
    di2 = np.array([random.randint(1,6) for x in range(1001)])
    di3 = np.array([random.randint(1,6) for x in range(1001)])

    di = np.array([di1,di2,di3]).T
    di.shape
    p = ((di == 4).sum(axis = 1) == 2).sum() / di.shape[0]
    prob.append(p)
```

```python
sns.distplot(prob)
```
