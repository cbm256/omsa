```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy.ndimage import gaussian_filter, laplace, gaussian_laplace

%matplotlib inline
```

```python
eye = np.array(Image.open('eye.jpg').convert('L'))
camera = np.array(Image.open('Camera.jpg').convert('L'))
```

```python
bins = np.arange(0, 1, .001)
fig, (ax1, ax2) = plt.subplots(1, 2)
p1 = ax1.hist((eye.ravel()/255), bins=bins, range=(0.0, 1.0), fc='k', ec='k')
p2 = ax2.hist((camera.ravel()/255), bins=bins, range=(0.0, 1.0), fc='k', ec='k')
```

```python
# part a

eye_filtered = gaussian_filter(eye, sigma=1)

# part b

camera_filtered = gaussian_laplace(camera, sigma=1)

# part c

hybrid = np.rint((eye_filtered + camera_filtered) / 2)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20,10))

p1 = ax1.imshow(eye, vmin=0, vmax=255, cmap='gray')
p2 = ax2.imshow(camera, vmin=0, vmax=255, cmap='gray')
p4 = ax3.imshow(eye_filtered, vmin=0, vmax=255, cmap='gray')
p5 = ax4.imshow(camera_filtered, vmin=0, vmax=255, cmap='gray')

```

The above images show the original (upper) and filtered (lower) eye and camera photos.
